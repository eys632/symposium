## 1. ÏµúÏ†ÄÏ†ê Íµ¨ÌïòÍ∏∞
$$ y = 2sin(x) + 0.5x , -5<=x<=13 $$

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
import math

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (ÌïúÍ∏Ä Íπ®Ïßê Î∞©ÏßÄ)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

# WindowsÏóêÏÑú ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÇ¨Ïö© (ÏãúÏä§ÌÖúÏóê Îî∞Îùº ÏûêÎèô ÏÑ†ÌÉù)
import platform
if platform.system() == 'Windows':
    try:
        # Windows ÌïúÍ∏Ä Ìè∞Ìä∏ ÏãúÎèÑ
        plt.rcParams['font.family'] = 'Malgun Gothic'
    except:
        try:
            plt.rcParams['font.family'] = 'AppleGothic'
        except:
            # Í∏∞Î≥∏ ÌïúÍ∏Ä ÏßÄÏõê Ìè∞Ìä∏Î°ú ÏÑ§Ï†ï
            import matplotlib.font_manager as fm
            font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
            
            # ÌïúÍ∏Ä Ìè∞Ìä∏ Ï∞æÍ∏∞
            korean_fonts = []
            for font in font_list:
                try:
                    font_name = fm.FontProperties(fname=font).get_name()
                    if any(keyword in font_name.lower() for keyword in ['malgun', 'batang', 'dotum', 'gulim', 'arial']):
                        korean_fonts.append(font_name)
                except:
                    continue
            
            if korean_fonts:
                plt.rcParams['font.family'] = korean_fonts[0]
            else:
                # ÎßàÏßÄÎßâ ÏàòÎã®: ÏãúÏä§ÌÖú Í∏∞Î≥∏ Ìè∞Ìä∏
                plt.rcParams['font.family'] = ['sans-serif']
                
# ÌïúÍ∏Ä Ìè∞Ìä∏ ÌôïÏù∏
print(f"ÌòÑÏû¨ ÏÑ§Ï†ïÎêú Ìè∞Ìä∏: {plt.rcParams['font.family']}")

# ÏùåÏàò Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
plt.rcParams['axes.unicode_minus'] = False

# x Î≤îÏúÑÎ•º 0.001 Í∞ÑÍ≤©ÏúºÎ°ú Ï¥òÏ¥òÌïòÍ≤å ÏÑ§Ï†ï
x = np.arange(-5, 13, 0.00001)

# Ï£ºÏñ¥ÏßÑ Ìï®Ïàò y = 2sin(x) + 0.5x Í≥ÑÏÇ∞
y = 2 * np.sin(x) + 0.5 * x

# ÏµúÏÜüÍ∞íÍ≥º Ìï¥Îãπ x Í∞í Ï∞æÍ∏∞
min_index = np.argmin(y)
min_x = x[min_index]
min_y = y[min_index]

# Ìï®ÏàòÏùò ÎØ∏Î∂ÑÏùÑ Ïù¥Ïö©Ìïú ÏûÑÍ≥ÑÏ†ê Î∂ÑÏÑù

# Ìï®Ïàò Ï†ïÏùò
def f(x):
    return 2 * np.sin(x) + 0.5 * x

# 1Ï∞® ÎØ∏Î∂Ñ Ìï®Ïàò (ÎèÑÌï®Ïàò)
def f_prime(x):
    return 2 * np.cos(x) + 0.5

# ÏàòÏπòÏ†Å ÏµúÏÜüÍ∞í Ï∞æÍ∏∞ (Îçî Ï†ïÌôïÌïú Î∞©Î≤ï)
result = minimize_scalar(f, bounds=(-5, 13), method='bounded')
exact_min_x = result.x
exact_min_y = result.fun

# ÏûÑÍ≥ÑÏ†ê Ï∞æÍ∏∞ (1Ï∞® ÎØ∏Î∂Ñ = 0Ïù∏ Ï†êÎì§)
# 2cos(x) + 0.5 = 0
# cos(x) = -0.25
critical_angles = []
x_test = np.linspace(-5, 13, 10000)
for i, x_val in enumerate(x_test[:-1]):
    if f_prime(x_val) * f_prime(x_test[i+1]) < 0:  # Î∂ÄÌò∏Í∞Ä Î∞îÎÄåÎäî ÏßÄÏ†ê
        critical_angles.append(x_val)

# ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
import matplotlib.pyplot as plt

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Í∑∏ÎûòÌîÑÏóêÏÑú ÌïúÍ∏Ä Íπ®Ïßê Î∞©ÏßÄ)
ensure_korean_font()

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(12, 8))

# Ìï®Ïàò Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')

# ÏµúÏÜüÍ∞í ÏßÄÏ†ê ÌëúÏãú
plt.plot(min_x, min_y, 'ro', markersize=10, label=f'ÏµúÏÜüÍ∞í: ({min_x:.3f}, {min_y:.3f})')

# ÏµúÏÜüÍ∞íÏóêÏÑú ÏàòÏßÅÏÑ†Í≥º ÏàòÌèâÏÑ† Í∑∏Î¶¨Í∏∞
plt.axvline(x=min_x, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=min_y, color='red', linestyle='--', alpha=0.7)

# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Ìï®Ïàò y = 2sin(x) + 0.5xÏùò Í∑∏ÎûòÌîÑÏôÄ ÏµúÏÜüÍ∞í', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

# xÏ∂ïÍ≥º yÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
plt.xlim(-5, 13)
plt.ylim(min_y - 1, max(y) + 1)

# ÏµúÏÜüÍ∞í Ï†ïÎ≥¥Î•º ÌÖçÏä§Ìä∏Î°ú ÌëúÏãú
plt.text(min_x + 0.5, min_y + 0.5, 
         f'ÏµúÏÜüÍ∞í\nx = {min_x:.4f}\ny = {min_y:.4f}', 
         fontsize=10, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

plt.tight_layout()
plt.show()

# ÏàòÌïôÏ†Å Î∂ÑÏÑù Í≤∞Í≥º
print(f"\n=== ÏàòÌïôÏ†Å Î∂ÑÏÑù Í≤∞Í≥º ===")
print(f"‚Ä¢ Ìï®Ïàò: y = 2sin(x) + 0.5x")
print(f"‚Ä¢ 1Ï∞® ÎØ∏Î∂Ñ: y' = 2cos(x) + 0.5")
print(f"‚Ä¢ ÏûÑÍ≥ÑÏ†ê Ï°∞Í±¥: 2cos(x) + 0.5 = 0 ‚Üí cos(x) = -0.25")
print(f"‚Ä¢ cos(x) = -0.25Ïùò Ìï¥: x = ¬±arccos(0.25) + 2œÄn")
print(f"‚Ä¢ Ï£ºÏñ¥ÏßÑ Íµ¨Í∞Ñ [-5, 13]ÏóêÏÑú ÏûÑÍ≥ÑÏ†ê Í∞úÏàò: {len(critical_angles)}Í∞ú")
print(f"‚Ä¢ Ï†ÑÏó≠ ÏµúÏÜüÍ∞í: x = {exact_min_x:.6f}, y = {exact_min_y:.6f}")

# arccos(-0.25) Í≥ÑÏÇ∞
arccos_val = math.acos(0.25)
print(f"‚Ä¢ arccos(0.25) = {arccos_val:.6f} (ÏïΩ {math.degrees(arccos_val):.2f}ÎèÑ)")
print(f"‚Ä¢ ÏûÑÍ≥ÑÏ†êÎì§ÏùÄ œÄ - arccos(0.25) + 2œÄn, œÄ + arccos(0.25) + 2œÄn ÌòïÌÉú")

# ÏÉÅÏÑ∏Ìïú ÏãúÍ∞ÅÌôî
plt.figure(figsize=(15, 10))

# Ï≤´ Î≤àÏß∏ ÏÑúÎ∏åÌîåÎ°Ø: ÏõêÌï®Ïàò
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')
plt.plot(exact_min_x, exact_min_y, 'ro', markersize=10, label='ÏµúÏÜüÍ∞í')
for angle in critical_angles:
    plt.plot(angle, f(angle), 'go', markersize=8)
    
plt.xlabel('x')
plt.ylabel('y')
plt.title('ÏõêÌï®Ïàò: y = 2sin(x) + 0.5x')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

# Îëê Î≤àÏß∏ ÏÑúÎ∏åÌîåÎ°Ø: 1Ï∞® ÎØ∏Î∂Ñ
plt.subplot(2, 2, 2)
y_prime = f_prime(x)
plt.plot(x, y_prime, 'g-', linewidth=2, label="y' = 2cos(x) + 0.5")
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='y = 0')
for angle in critical_angles:
    plt.plot(angle, 0, 'go', markersize=8)
    
plt.xlabel('x')
plt.ylabel("y'")
plt.title('1Ï∞® ÎØ∏Î∂Ñ: Í∏∞Ïö∏Í∏∞Ïùò Î≥ÄÌôî')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

# ÏÑ∏ Î≤àÏß∏ ÏÑúÎ∏åÌîåÎ°Ø: ÌôïÎåÄÎêú ÏµúÏÜüÍ∞í Í∑ºÏ≤ò
plt.subplot(2, 2, 3)
x_zoom = np.linspace(exact_min_x - 2, exact_min_x + 2, 1000)
y_zoom = f(x_zoom)
plt.plot(x_zoom, y_zoom, 'b-', linewidth=2)
plt.plot(exact_min_x, exact_min_y, 'ro', markersize=10)
plt.axvline(x=exact_min_x, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=exact_min_y, color='red', linestyle='--', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'ÏµúÏÜüÍ∞í Í∑ºÏ≤ò ÌôïÎåÄ\n(x={exact_min_x:.4f}, y={exact_min_y:.4f})')
plt.grid(True, alpha=0.3)

# ÎÑ§ Î≤àÏß∏ ÏÑúÎ∏åÌîåÎ°Ø: Ìï®ÏàòÏùò Íµ¨ÏÑ± ÏöîÏÜå Î∂ÑÌï¥
plt.subplot(2, 2, 4)
sin_component = 2 * np.sin(x)
linear_component = 0.5 * x
plt.plot(x, sin_component, 'r--', linewidth=2, alpha=0.7, label='2sin(x)')
plt.plot(x, linear_component, 'g--', linewidth=2, alpha=0.7, label='0.5x')
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ìï®Ïàò Íµ¨ÏÑ± ÏöîÏÜå Î∂ÑÌï¥')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

plt.tight_layout()
plt.show()

# ÏµúÏÜüÍ∞í Ï†ïÎ≥¥ Ï∂úÎ†•
print(f"=== Ìï®Ïàò y = 2sin(x) + 0.5xÏùò ÏµúÏÜüÍ∞í Î∂ÑÏÑù ===")
print(f"Íµ¨Í∞Ñ: [-5, 13]")
print(f"ÏµúÏÜüÍ∞íÏù¥ Î∞úÏÉùÌïòÎäî x: {min_x:.6f}")
print(f"ÏµúÏÜüÍ∞í: {min_y:.6f}")
print(f"ÏµúÏÜüÍ∞í ÏúÑÏπò: ({min_x:.4f}, {min_y:.4f})")

# Ï†ÑÏ≤¥ Íµ¨Í∞ÑÏóêÏÑú Î™®Îì† Í∑πÍ∞íÎì§ÏùÑ Ï∞æÏïÑ ÎπÑÍµê
from scipy.optimize import fminbound
import matplotlib.pyplot as plt

print("=== Ï†ÑÏ≤¥ Íµ¨Í∞Ñ [-5, 13]ÏóêÏÑú Î™®Îì† Í∑πÍ∞í Î∂ÑÏÑù ===")

# Î™®Îì† ÏûÑÍ≥ÑÏ†êÏóêÏÑúÏùò Ìï®ÏàòÍ∞í Í≥ÑÏÇ∞
all_critical_points = []

# 1Ï∞® ÎØ∏Î∂ÑÏù¥ 0Ïù¥ ÎêòÎäî Ï†êÎì§ÏùÑ Îçî Ï†ïÌôïÌûà Ï∞æÍ∏∞
x_fine = np.linspace(-5, 13, 50000)
y_prime_fine = 2 * np.cos(x_fine) + 0.5

# Î∂ÄÌò∏Í∞Ä Î∞îÎÄåÎäî ÏßÄÏ†êÎì§ Ï∞æÍ∏∞
for i in range(len(x_fine)-1):
    if y_prime_fine[i] * y_prime_fine[i+1] < 0:
        # Îçî Ï†ïÌôïÌïú Í∞íÏùÑ ÏúÑÌï¥ Ìï¥Îãπ Íµ¨Í∞ÑÏóêÏÑú ÏÑ∏Î∂ÑÌôî
        x_local = np.linspace(x_fine[i], x_fine[i+1], 1000)
        y_prime_local = 2 * np.cos(x_local) + 0.5
        zero_idx = np.argmin(np.abs(y_prime_local))
        critical_x = x_local[zero_idx]
        critical_y = f(critical_x)
        all_critical_points.append((critical_x, critical_y))

# Íµ¨Í∞Ñ ÎÅùÏ†êÎì§ÎèÑ Ìè¨Ìï®
all_critical_points.append((-5, f(-5)))
all_critical_points.append((13, f(13)))

# Ï†ïÎ†¨ Î∞è Ï∂úÎ†•
all_critical_points.sort()

print(f"Î∞úÍ≤¨Îêú Î™®Îì† ÏûÑÍ≥ÑÏ†ê Î∞è ÎÅùÏ†êÎì§:")
for i, (x_val, y_val) in enumerate(all_critical_points):
    point_type = "ÎÅùÏ†ê" if x_val == -5 or x_val == 13 else "ÏûÑÍ≥ÑÏ†ê"
    print(f"{i+1}. x = {x_val:8.4f}, y = {y_val:8.4f} ({point_type})")

# Ï†ÑÏó≠ ÏµúÏÜüÍ∞íÍ≥º ÏµúÎåìÍ∞í Ï∞æÍ∏∞
min_point = min(all_critical_points, key=lambda p: p[1])
max_point = max(all_critical_points, key=lambda p: p[1])

print(f"\n=== ÏµúÏ¢Ö Í≤∞Í≥º ===")
print(f"üîµ Ï†ÑÏó≠ ÏµúÏÜüÍ∞í: x = {min_point[0]:.6f}, y = {min_point[1]:.6f}")
print(f"üî¥ Ï†ÑÏó≠ ÏµúÎåìÍ∞í: x = {max_point[0]:.6f}, y = {max_point[1]:.6f}")

# ÏãúÍ∞ÅÌôî Í∞úÏÑ†
plt.figure(figsize=(14, 8))

# Ìï®Ïàò Í∑∏ÎûòÌîÑ
x_plot = np.linspace(-5, 13, 2000)
y_plot = 2 * np.sin(x_plot) + 0.5 * x_plot

plt.plot(x_plot, y_plot, 'b-', linewidth=2.5, label='y = 2sin(x) + 0.5x')

# Î™®Îì† ÏûÑÍ≥ÑÏ†ê ÌëúÏãú
for x_val, y_val in all_critical_points:
    if x_val == -5 or x_val == 13:
        plt.plot(x_val, y_val, 'ks', markersize=8, label='ÎÅùÏ†ê' if x_val == -5 else '')
    else:
        plt.plot(x_val, y_val, 'go', markersize=8, label='ÏûÑÍ≥ÑÏ†ê' if x_val == all_critical_points[1][0] else '')

# Ï†ÑÏó≠ ÏµúÏÜüÍ∞íÍ≥º ÏµúÎåìÍ∞í Í∞ïÏ°∞
plt.plot(min_point[0], min_point[1], 'ro', markersize=12, 
         label=f'Ï†ÑÏó≠ ÏµúÏÜüÍ∞í: ({min_point[0]:.3f}, {min_point[1]:.3f})')
plt.plot(max_point[0], max_point[1], 'mo', markersize=12, 
         label=f'Ï†ÑÏó≠ ÏµúÎåìÍ∞í: ({max_point[0]:.3f}, {max_point[1]:.3f})')

# ÏµúÏÜüÍ∞íÏóêÏÑú ÏÑ† Í∑∏Î¶¨Í∏∞
plt.axvline(x=min_point[0], color='red', linestyle='--', alpha=0.7)
plt.axhline(y=min_point[1], color='red', linestyle='--', alpha=0.7)

# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Ìï®Ïàò y = 2sin(x) + 0.5x: Î™®Îì† Í∑πÍ∞í Î∂ÑÏÑù', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)
plt.xlim(-5, 13)

# ÏµúÏÜüÍ∞í Ï†ïÎ≥¥ ÌÖçÏä§Ìä∏
plt.text(min_point[0] + 0.5, min_point[1] + 0.8, 
         f'Ï†ÑÏó≠ ÏµúÏÜüÍ∞í\nx = {min_point[0]:.4f}\ny = {min_point[1]:.4f}', 
         fontsize=11, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.4", facecolor="yellow", alpha=0.8))

plt.tight_layout()
plt.show()

# ÏàòÌïôÏ†Å Î∂ÑÏÑù
print(f"\n=== ÏàòÌïôÏ†Å Ìï¥ÏÑù ===")
print(f"‚Ä¢ Ìï®Ïàò y = 2sin(x) + 0.5xÎäî Ï£ºÍ∏∞Ï†Å ÏÑ±Î∂ÑÍ≥º ÏÑ†Ìòï ÏÑ±Î∂ÑÏùò Ìï©")
print(f"‚Ä¢ ÏÇ¨Ïù∏ Ìï®ÏàòÏùò ÏßÑÌè≠: 2")
print(f"‚Ä¢ ÏÑ†Ìòï Ìï®ÏàòÏùò Í∏∞Ïö∏Í∏∞: 0.5")
print(f"‚Ä¢ 1Ï∞® ÎØ∏Î∂Ñ: y' = 2cos(x) + 0.5")
print(f"‚Ä¢ ÏûÑÍ≥ÑÏ†ê Ï°∞Í±¥: cos(x) = -0.25")
print(f"‚Ä¢ cos(x) = -0.25Îäî ÏïΩ {np.degrees(np.arccos(0.25)):.1f}¬∞ÏôÄ {360-np.degrees(np.arccos(0.25)):.1f}¬∞ÏóêÏÑú Î∞úÏÉù")
print(f"‚Ä¢ ÏÑ†Ìòï ÏÑ±Î∂Ñ ÎïåÎ¨∏Ïóê Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú Ïö∞ÏÉÅÌñ•ÌïòÎäî Í≤ΩÌñ•")
print(f"‚Ä¢ Îî∞ÎùºÏÑú Íµ¨Í∞ÑÏùò ÏôºÏ™Ω Î∂ÄÎ∂ÑÏóêÏÑú ÏµúÏÜüÍ∞íÏù¥ Î∞úÏÉù")

# ÌïúÍ∏Ä Ìè∞Ìä∏ Î¨∏Ï†ú Ìï¥Í≤∞ÏùÑ ÏúÑÌïú Ï∂îÍ∞Ä ÏÑ§Ï†ï
def setup_korean_font():
    """ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï Ìï®Ïàò"""
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm
    import platform
    
    # ÏãúÏä§ÌÖúÎ≥Ñ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
    if platform.system() == 'Windows':
        # Windows ÌïúÍ∏Ä Ìè∞Ìä∏Îì§
        fonts = ['Malgun Gothic', 'Batang', 'Dotum', 'Gulim', 'Arial Unicode MS']
    elif platform.system() == 'Darwin':  # macOS
        fonts = ['AppleGothic', 'Arial Unicode MS']
    else:  # Linux
        fonts = ['DejaVu Sans', 'Liberation Sans']
    
    # ÏÇ¨Ïö© Í∞ÄÎä•Ìïú Ìè∞Ìä∏ Ï∞æÍ∏∞
    available_fonts = [f.name for f in fm.fontManager.ttflist]
    
    for font in fonts:
        if font in available_fonts:
            plt.rcParams['font.family'] = font
            print(f"‚úÖ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï ÏôÑÎ£å: {font}")
            break
    else:
        # ÌïúÍ∏Ä Ìè∞Ìä∏Î•º Ï∞æÏßÄ Î™ªÌïú Í≤ΩÏö∞
        print("‚ö†Ô∏è ÌïúÍ∏Ä Ìè∞Ìä∏Î•º Ï∞æÏßÄ Î™ªÌñàÏäµÎãàÎã§. Í∏∞Î≥∏ ÏÑ§Ï†ïÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.")
        plt.rcParams['font.family'] = ['sans-serif']
    
    # ÏùåÏàò Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
    plt.rcParams['axes.unicode_minus'] = False
    
    return plt.rcParams['font.family']

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï Ïã§Ìñâ
current_font = setup_korean_font()

# Ìè∞Ìä∏ ÌÖåÏä§Ìä∏
plt.figure(figsize=(8, 4))
plt.text(0.5, 0.5, 'ÌïúÍ∏Ä Ìè∞Ìä∏ ÌÖåÏä§Ìä∏: Í∞ÄÎÇòÎã§ÎùºÎßàÎ∞îÏÇ¨ÏïÑÏ∞®Ïπ¥ÌÉÄÌååÌïò', 
         ha='center', va='center', fontsize=16)
plt.text(0.5, 0.3, f'ÌòÑÏû¨ Ìè∞Ìä∏: {current_font}', 
         ha='center', va='center', fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title('ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï ÌôïÏù∏')
plt.axis('off')
plt.tight_layout()
plt.show()
````

## 2. Ïà®Í≤®ÏßÑ ÏµúÏÜüÍ∞í Ï∞æÍ∏∞

$$ f(x,y) = cos(x) + cos(y) $$     
$$-2{\pi} <= x <= 2{\pi},$$
$$-2{\pi} <= y <= 2{\pi} $$

````python
### 2Î≤à Î¨∏Ï†ú Ìï¥Í≤∞ Ï†ëÍ∑º Î∞©Î≤ïÎì§

**Î¨∏Ï†ú**: $f(x,y) = \cos(x) + \cos(y)$ÏóêÏÑú $-2\pi \leq x \leq 2\pi$, $-2\pi \leq y \leq 2\pi$ Î≤îÏúÑÏùò ÏµúÏÜüÍ∞í Ï∞æÍ∏∞

#### üéØ **Î∞©Î≤ï 1: Ìï¥ÏÑùÏ†Å Ï†ëÍ∑º (Analytical Method)**
- **Ìé∏ÎØ∏Î∂ÑÏùÑ Ïù¥Ïö©Ìïú ÏûÑÍ≥ÑÏ†ê Ï∞æÍ∏∞**
  - $\frac{\partial f}{\partial x} = -\sin(x) = 0$ ‚Üí $x = n\pi$ (Ï†ïÏàò n)
  - $\frac{\partial f}{\partial y} = -\sin(y) = 0$ ‚Üí $y = n\pi$ (Ï†ïÏàò n)
  - Ï£ºÏñ¥ÏßÑ Íµ¨Í∞ÑÏóêÏÑú Í∞ÄÎä•Ìïú ÏûÑÍ≥ÑÏ†êÎì§: $(0,0), (\pi,0), (0,\pi), (\pi,\pi)$ Îì±
- **Í≤ΩÍ≥ÑÍ∞í Í≤ÄÏÇ¨**
- **2Ï∞® ÎØ∏Î∂Ñ ÌåêÏ†ïÎ≤ïÏúºÎ°ú Í∑πÍ∞í Î∂ÑÎ•ò**

#### üî¢ **Î∞©Î≤ï 2: ÏàòÏπòÏ†Å ÏµúÏ†ÅÌôî (Numerical Optimization)**
- **scipy.optimize.minimize** ÏÇ¨Ïö©
- **Global optimization**: `scipy.optimize.differential_evolution`
- **Grid search + local optimization**

#### üìä **Î∞©Î≤ï 3: Í≤©ÏûêÏ†ê ÌÉêÏÉâ (Grid Search)**
- **Í≥†Ìï¥ÏÉÅÎèÑ Í≤©ÏûêÎ°ú Ï†ÑÏ≤¥ ÏòÅÏó≠ Ïä§Ï∫î**
- **NumPy meshgrid ÌôúÏö©**
- **Brute force Î∞©ÏãùÏúºÎ°ú Î™®Îì† Ï†ê Í≤ÄÏÇ¨**

#### üé® **Î∞©Î≤ï 4: ÏãúÍ∞ÅÌôî Í∏∞Î∞ò Ï†ëÍ∑º (Visualization)**
- **3D Surface Plot**
- **Contour Plot (Îì±Í≥†ÏÑ† Í∑∏ÎûòÌîÑ)**
- **Heatmap**
- **ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÏµúÏÜüÍ∞í ÏúÑÏπò ÌôïÏù∏ ÌõÑ Ï†ïÎ∞Ä Í≥ÑÏÇ∞**

#### üßÆ **Î∞©Î≤ï 5: ÎåÄÏπ≠ÏÑ± ÌôúÏö© (Symmetry Analysis)**
- **Ìï®ÏàòÏùò Ï£ºÍ∏∞ÏÑ±Í≥º ÎåÄÏπ≠ÏÑ± Ïù¥Ïö©**
- $\cos(x)$Îäî $2\pi$ Ï£ºÍ∏∞, Ïö∞Ìï®Ïàò
- $f(x,y) = f(-x,y) = f(x,-y) = f(-x,-y)$
- Ï≤´ Î≤àÏß∏ ÏÇ¨Î∂ÑÎ©¥Îßå Î∂ÑÏÑù ÌõÑ ÌôïÏû•

#### üî¨ **Î∞©Î≤ï 6: ÏàòÌïôÏ†Å Î∂ÑÏÑù (Mathematical Analysis)**
- **ÏΩîÏÇ¨Ïù∏ Ìï®ÏàòÏùò ÏÑ±Ïßà ÌôúÏö©**
- $\cos(x)$Ïùò ÏµúÏÜüÍ∞íÏùÄ $x = \pi + 2n\pi$ÏóêÏÑú -1
- $f(x,y) = \cos(x) + \cos(y)$Ïùò ÏµúÏÜüÍ∞íÏùÄ $-1 + (-1) = -2$
- ÏµúÏÜüÍ∞í ÏúÑÏπò: $(x,y) = (\pi, \pi), (-\pi, \pi), (\pi, -\pi), (-\pi, -\pi)$

#### üöÄ **Í∞Å Î∞©Î≤ïÏùò Ïû•Îã®Ï†ê**

| Î∞©Î≤ï | Ïû•Ï†ê | Îã®Ï†ê | Ï†ïÌôïÎèÑ |
|------|------|------|--------|
| Ìï¥ÏÑùÏ†Å | Ï†ïÌôïÌïú Ìï¥, ÏàòÌïôÏ†Å Ïù¥Ìï¥ | Î≥µÏû°Ìïú Ìï®ÏàòÏóê ÌïúÍ≥Ñ | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê |
| ÏàòÏπòÏ†Å ÏµúÏ†ÅÌôî | Î≤îÏö©ÏÑ±, Îπ†Î¶Ñ | ÏßÄÏó≠ ÏµúÏÜüÍ∞í ÏúÑÌóò | ‚≠ê‚≠ê‚≠ê‚≠ê |
| Í≤©ÏûêÏ†ê ÌÉêÏÉâ | ÌôïÏã§Ìï®, ÏßÅÍ¥ÄÏ†Å | Í≥ÑÏÇ∞Îüâ ÎßéÏùå | ‚≠ê‚≠ê‚≠ê‚≠ê |
| ÏãúÍ∞ÅÌôî | ÏßÅÍ¥ÄÏ†Å Ïù¥Ìï¥ | Í∑ºÏÇ¨Ïπò | ‚≠ê‚≠ê‚≠ê |
| ÎåÄÏπ≠ÏÑ± ÌôúÏö© | Ìö®Ïú®Ï†Å | ÌäπÏàòÌïú Í≤ΩÏö∞Îßå | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê |
| ÏàòÌïôÏ†Å Î∂ÑÏÑù | Í∞ÄÏû• Ï†ïÌôï | Ìï®ÏàòÎ≥Ñ Î∂ÑÏÑù ÌïÑÏöî | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê |

#### üí° **Í∂åÏû• Ï†ëÍ∑º ÏàúÏÑú**
1. **ÏàòÌïôÏ†Å Î∂ÑÏÑù**ÏúºÎ°ú Ïù¥Î°†Ï†Å Îãµ Íµ¨ÌïòÍ∏∞
2. **ÏãúÍ∞ÅÌôî**Î°ú Ìï®Ïàò ÌòïÌÉú ÌååÏïÖ
3. **Ìï¥ÏÑùÏ†Å Î∞©Î≤ï**ÏúºÎ°ú Ï†ïÌôïÌïú ÏûÑÍ≥ÑÏ†ê Í≥ÑÏÇ∞  
4. **ÏàòÏπòÏ†Å Î∞©Î≤ï**ÏúºÎ°ú Í≤ÄÏ¶ù
5. **Î™®Îì† Î∞©Î≤ï Í≤∞Í≥º ÎπÑÍµê**
````

````python
# 2Î≤à Î¨∏Ï†ú: f(x,y) = cos(x) + cos(y) ÏµúÏÜüÍ∞í Ï∞æÍ∏∞
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution
from mpl_toolkits.mplot3d import Axes3D

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
ensure_korean_font()

print("üéØ Î¨∏Ï†ú 2: f(x,y) = cos(x) + cos(y)Ïùò ÏµúÏÜüÍ∞í Ï∞æÍ∏∞")
print("üìê Ï†ïÏùòÏó≠: -2œÄ ‚â§ x ‚â§ 2œÄ, -2œÄ ‚â§ y ‚â§ 2œÄ")
print("="*60)

# Ìï®Ïàò Ï†ïÏùò
def f(x, y):
    """Î™©Ï†Å Ìï®Ïàò: f(x,y) = cos(x) + cos(y)"""
    return np.cos(x) + np.cos(y)

def f_1d(vars):
    """1Ï∞®Ïõê Î∞∞Ïó¥ ÏûÖÎ†•Ïö© Ìï®Ïàò (ÏµúÏ†ÅÌôî ÏïåÍ≥†Î¶¨Ï¶òÏö©)"""
    return f(vars[0], vars[1])

# Ï†ïÏùòÏó≠ ÏÑ§Ï†ï
x_min, x_max = -2*np.pi, 2*np.pi
y_min, y_max = -2*np.pi, 2*np.pi

print("\n" + "="*60)
print("üßÆ Î∞©Î≤ï 1: ÏàòÌïôÏ†Å Î∂ÑÏÑù (Mathematical Analysis)")
print("="*60)

print("\nüìä ÏΩîÏÇ¨Ïù∏ Ìï®ÏàòÏùò ÏÑ±Ïßà Î∂ÑÏÑù:")
print("‚Ä¢ cos(x)Ïùò ÏµúÏÜüÍ∞í: -1 (x = œÄ + 2nœÄÏóêÏÑú)")
print("‚Ä¢ cos(x)Ïùò ÏµúÎåìÍ∞í: +1 (x = 2nœÄÏóêÏÑú)")
print("‚Ä¢ f(x,y) = cos(x) + cos(y)Ïù¥ÎØÄÎ°ú:")
print("  - ÏµúÏÜüÍ∞í: (-1) + (-1) = -2")
print("  - ÏµúÎåìÍ∞í: (+1) + (+1) = +2")

print("\nüéØ ÏµúÏÜüÍ∞íÏù¥ Î∞úÏÉùÌïòÎäî Ï°∞Í±¥:")
print("‚Ä¢ cos(x) = -1Ïù¥Í≥† cos(y) = -1Ïùº Îïå")
print("‚Ä¢ x = œÄ + 2nœÄ, y = œÄ + 2mœÄ (n, mÏùÄ Ï†ïÏàò)")

print("\nüìç Ï£ºÏñ¥ÏßÑ Íµ¨Í∞Ñ [-2œÄ, 2œÄ]ÏóêÏÑú ÏµúÏÜüÍ∞í ÏúÑÏπò:")
theoretical_min_points = []
for x_val in [-np.pi, np.pi]:
    for y_val in [-np.pi, np.pi]:
        if x_min <= x_val <= x_max and y_min <= y_val <= y_max:
            theoretical_min_points.append((x_val, y_val))
            print(f"  ‚Ä¢ ({x_val/np.pi:.0f}œÄ, {y_val/np.pi:.0f}œÄ) = ({x_val:.4f}, {y_val:.4f})")
            print(f"    f({x_val:.4f}, {y_val:.4f}) = {f(x_val, y_val):.6f}")

theoretical_min_value = -2
print(f"\n‚úÖ Ïù¥Î°†Ï†Å ÏµúÏÜüÍ∞í: {theoretical_min_value}")

print("\n" + "="*60)
print("üî¢ Î∞©Î≤ï 2: ÏàòÏπòÏ†Å ÏµúÏ†ÅÌôî (Numerical Optimization)")
print("="*60)

# Î∞©Î≤ï 2-1: ÏßÄÏó≠ ÏµúÏ†ÅÌôî (Ïó¨Îü¨ ÏãúÏûëÏ†êÏóêÏÑú)
print("\nüéØ Î∞©Î≤ï 2-1: Ïó¨Îü¨ ÏãúÏûëÏ†êÏóêÏÑú ÏßÄÏó≠ ÏµúÏ†ÅÌôî")
local_results = []
start_points = [
    (0, 0), (np.pi, 0), (0, np.pi), (np.pi, np.pi),
    (-np.pi, 0), (0, -np.pi), (-np.pi, -np.pi),
    (np.pi/2, np.pi/2), (-np.pi/2, -np.pi/2)
]

for i, (x0, y0) in enumerate(start_points):
    result = minimize(f_1d, [x0, y0], 
                     bounds=[(x_min, x_max), (y_min, y_max)],
                     method='L-BFGS-B')
    local_results.append(result)
    print(f"ÏãúÏûëÏ†ê {i+1}: ({x0:.2f}, {y0:.2f}) ‚Üí "
          f"ÏµúÏÜüÍ∞í {result.fun:.6f}, ÏúÑÏπò: ({result.x[0]:.4f}, {result.x[1]:.4f})")

# ÏµúÏÑ†Ïùò ÏßÄÏó≠ ÏµúÏ†ÅÌôî Í≤∞Í≥º
best_local = min(local_results, key=lambda r: r.fun)
print(f"\nüèÜ ÏµúÏÑ†Ïùò ÏßÄÏó≠ ÏµúÏ†ÅÌôî Í≤∞Í≥º: {best_local.fun:.6f}")
print(f"   ÏúÑÏπò: ({best_local.x[0]:.4f}, {best_local.x[1]:.4f})")

# Î∞©Î≤ï 2-2: Ï†ÑÏó≠ ÏµúÏ†ÅÌôî
print("\nüåê Î∞©Î≤ï 2-2: Ï†ÑÏó≠ ÏµúÏ†ÅÌôî (Differential Evolution)")
bounds = [(x_min, x_max), (y_min, y_max)]
global_result = differential_evolution(f_1d, bounds, seed=42, maxiter=1000)
print(f"Ï†ÑÏó≠ ÏµúÏ†ÅÌôî Í≤∞Í≥º: {global_result.fun:.6f}")
print(f"ÏúÑÏπò: ({global_result.x[0]:.4f}, {global_result.x[1]:.4f})")

print("\n" + "="*60)
print("üìä Î∞©Î≤ï 3: Í≥†Ìï¥ÏÉÅÎèÑ Í≤©Ïûê ÌÉêÏÉâ (High-Resolution Grid Search)")
print("="*60)

# Í≥†Ìï¥ÏÉÅÎèÑ Í≤©Ïûê ÏÉùÏÑ±
resolution = 1000  # Í≤©Ïûê Ìï¥ÏÉÅÎèÑ
x_grid = np.linspace(x_min, x_max, resolution)
y_grid = np.linspace(y_min, y_max, resolution)
X, Y = np.meshgrid(x_grid, y_grid)
Z = f(X, Y)

# ÏµúÏÜüÍ∞íÍ≥º ÏúÑÏπò Ï∞æÍ∏∞
min_idx = np.unravel_index(np.argmin(Z), Z.shape)
grid_min_x = X[min_idx]
grid_min_y = Y[min_idx]
grid_min_value = Z[min_idx]

print(f"Í≤©Ïûê Ìï¥ÏÉÅÎèÑ: {resolution} √ó {resolution} = {resolution**2:,} Ï†ê")
print(f"Í≤©Ïûê ÌÉêÏÉâ ÏµúÏÜüÍ∞í: {grid_min_value:.6f}")
print(f"ÏúÑÏπò: ({grid_min_x:.4f}, {grid_min_y:.4f})")

# ÏµúÏÜüÍ∞í Í∑ºÏ≤òÏùò Îã§Î•∏ Ï†êÎì§ÎèÑ ÌôïÏù∏
min_threshold = grid_min_value + 1e-10
min_mask = Z <= min_threshold
min_indices = np.where(min_mask)
print(f"\nÏµúÏÜüÍ∞í Í∑ºÏ≤ò Ï†êÎì§Ïùò Í∞úÏàò: {len(min_indices[0])}")

print("\n" + "="*60)
print("üìà Î∞©Î≤ï 4: Ìï¥ÏÑùÏ†Å Ï†ëÍ∑º (Critical Points Analysis)")
print("="*60)

print("\nüîç Ìé∏ÎØ∏Î∂ÑÏùÑ Ïù¥Ïö©Ìïú ÏûÑÍ≥ÑÏ†ê Ï∞æÍ∏∞:")
print("‚àÇf/‚àÇx = -sin(x) = 0  ‚Üí  x = nœÄ")
print("‚àÇf/‚àÇy = -sin(y) = 0  ‚Üí  y = mœÄ")

# Ï£ºÏñ¥ÏßÑ Íµ¨Í∞ÑÏóêÏÑú Í∞ÄÎä•Ìïú ÏûÑÍ≥ÑÏ†êÎì§
critical_points = []
for n in range(-2, 3):  # -2œÄÎ∂ÄÌÑ∞ 2œÄÍπåÏßÄ
    for m in range(-2, 3):
        x_crit = n * np.pi
        y_crit = m * np.pi
        if x_min <= x_crit <= x_max and y_min <= y_crit <= y_max:
            f_val = f(x_crit, y_crit)
            critical_points.append((x_crit, y_crit, f_val))

# ÏûÑÍ≥ÑÏ†êÎì§ Ï†ïÎ†¨ (Ìï®ÏàòÍ∞í Í∏∞Ï§Ä)
critical_points.sort(key=lambda p: p[2])

print(f"\nüìç Î∞úÍ≤¨Îêú ÏûÑÍ≥ÑÏ†êÎì§ (Ï¥ù {len(critical_points)}Í∞ú):")
for i, (x, y, fval) in enumerate(critical_points):
    point_type = "ÏµúÏÜüÍ∞í" if abs(fval - (-2)) < 1e-10 else "ÏµúÎåìÍ∞í" if abs(fval - 2) < 1e-10 else "ÏïàÏû•Ï†ê"
    print(f"{i+1:2d}. ({x/np.pi:+.0f}œÄ, {y/np.pi:+.0f}œÄ) = ({x:+.4f}, {y:+.4f}), "
          f"f = {fval:+.6f} ({point_type})")

# ÏµúÏÜüÍ∞íÎì§Îßå Îî∞Î°ú Ï∂îÏ∂ú
min_points = [p for p in critical_points if abs(p[2] - (-2)) < 1e-10]
print(f"\nüéØ ÏµúÏÜüÍ∞í ÏúÑÏπòÎì§ (Ï¥ù {len(min_points)}Í∞ú):")
for i, (x, y, fval) in enumerate(min_points):
    print(f"{i+1}. ({x/np.pi:+.0f}œÄ, {y/np.pi:+.0f}œÄ) = ({x:+.4f}, {y:+.4f})")

print("\n" + "="*60)
print("üìä Í≤∞Í≥º Ï¢ÖÌï© Î∞è Í≤ÄÏ¶ù")
print("="*60)

results_summary = {
    "Ïù¥Î°†Ï†Å Î∂ÑÏÑù": theoretical_min_value,
    "ÏµúÏÑ† ÏßÄÏó≠ ÏµúÏ†ÅÌôî": best_local.fun,
    "Ï†ÑÏó≠ ÏµúÏ†ÅÌôî": global_result.fun,
    "Í≤©Ïûê ÌÉêÏÉâ": grid_min_value,
    "Ìï¥ÏÑùÏ†Å ÏûÑÍ≥ÑÏ†ê": min_points[0][2] if min_points else None
}

print("\nüèÜ Í∞Å Î∞©Î≤ïÎ≥Ñ ÏµúÏÜüÍ∞í Í≤∞Í≥º:")
for method, value in results_summary.items():
    if value is not None:
        error = abs(value - theoretical_min_value)
        print(f"‚Ä¢ {method:15s}: {value:10.6f} (Ïò§Ï∞®: {error:.2e})")

print(f"\n‚úÖ ÏµúÏ¢Ö Îãµ: f(x,y) = cos(x) + cos(y)Ïùò ÏµúÏÜüÍ∞íÏùÄ {theoretical_min_value}")
print(f"üìç ÏµúÏÜüÍ∞í ÏúÑÏπò: (¬±œÄ, ¬±œÄ)Ïùò Ï°∞Ìï© (Ï¥ù 4Í∞ú Ï†ê)")
print(f"üéØ Ï†ïÌôïÌïú ÏúÑÏπòÎì§:")
for x, y, _ in min_points:
    print(f"   ‚Ä¢ ({x:+.4f}, {y:+.4f})")

# Ïò§Ï∞® Î∂ÑÏÑù
print(f"\nüî¨ ÏàòÏπòÏ†Å Î∞©Î≤ïÎì§Ïùò Ï†ïÌôïÎèÑ:")
numerical_methods = ["ÏµúÏÑ† ÏßÄÏó≠ ÏµúÏ†ÅÌôî", "Ï†ÑÏó≠ ÏµúÏ†ÅÌôî", "Í≤©Ïûê ÌÉêÏÉâ"]
for method in numerical_methods:
    if method in results_summary:
        error = abs(results_summary[method] - theoretical_min_value)
        accuracy = max(0, -np.log10(error)) if error > 0 else float('inf')
        print(f"‚Ä¢ {method}: Ïò§Ï∞® {error:.2e}, Ï†ïÌôïÎèÑ ~{accuracy:.1f} ÏûêÎ¶øÏàò")
````

## üß† ÏÇ¨Ïö©Ïûê ÏßÄÏ†ï Í∑∏Î¶¨Îìú ÌÉêÏÉâ ÏïåÍ≥†Î¶¨Ï¶ò

### ÏïåÍ≥†Î¶¨Ï¶ò ÌäπÏßï:
1. **Î∂àÍ∑†Îì± Í∞ÑÍ≤© ÌÉêÏÉâ**: Ï†ïÌï¥ÏßÑ Ìå®ÌÑ¥ÏúºÎ°ú Í∞ÑÍ≤©ÏùÑ Í±¥ÎÑàÎõ∞Î©∞ ÌÉêÏÉâ
2. **Ï†ÅÏùëÏ†Å ÏÑ∏Î∞ÄÌôî**: ÏµúÏÜüÍ∞í Î∞úÍ≤¨ Ïãú Ï£ºÎ≥Ä ÏòÅÏó≠ÏùÑ Îçî ÏÑ∏Î∞ÄÌïòÍ≤å ÌÉêÏÉâ
3. **Í≥ÑÏ∏µÏ†Å Ï†ëÍ∑º**: Í±∞Ïπú ÌÉêÏÉâ ‚Üí ÏÑ∏Î∞ÄÌïú ÌÉêÏÉâÏùò Îã§Îã®Í≥Ñ Í≥ºÏ†ï
4. **Ìö®Ïú®Ï†Å Ïª§Î≤ÑÎ¶¨ÏßÄ**: Ï†ÑÏ≤¥ ÏòÅÏó≠ÏùÑ Ìö®Í≥ºÏ†ÅÏúºÎ°ú Ïª§Î≤ÑÌïòÎ©¥ÏÑúÎèÑ Í≥ÑÏÇ∞Îüâ ÏµúÏ†ÅÌôî

Ïù¥ ÏïåÍ≥†Î¶¨Ï¶òÏùÄ ÏùºÎ∞òÏ†ÅÏù∏ Í∑†Îì± Í∑∏Î¶¨Îìú ÌÉêÏÉâÎ≥¥Îã§ Ìö®Ïú®Ï†ÅÏù¥Î©¥ÏÑúÎèÑ, Î≥µÏû°Ìïú Ìï®ÏàòÏùò Ïà®Í≤®ÏßÑ ÏµúÏÜüÍ∞íÏùÑ Ï∞æÏïÑÎÇ¥Îäî Îä•Î†•Ïù¥ Îõ∞Ïñ¥ÎÇ©ÎãàÎã§.

````python
import numpy as np
import matplotlib.pyplot as plt
from typing import Callable, Tuple, List, Dict
import time

def custom_grid_search(func: Callable, 
                      x_bounds: Tuple[float, float], 
                      y_bounds: Tuple[float, float],
                      skip_pattern: List[int] = [1, 2, 3, 2, 1],  # Í±¥ÎÑàÎõ∏ Í∞ÑÍ≤© Ìå®ÌÑ¥
                      initial_points: int = 50,  # Ï¥àÍ∏∞ ÌÉêÏÉâÏ†ê Ïàò
                      refinement_levels: int = 3,  # ÏÑ∏Î∞ÄÌôî Îã®Í≥Ñ Ïàò
                      refinement_factor: float = 0.2,  # ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠ ÎπÑÏú®
                      verbose: bool = True) -> Dict:
    """
    ÏÇ¨Ïö©Ïûê ÏßÄÏ†ï Í∑∏Î¶¨Îìú ÌÉêÏÉâ ÏïåÍ≥†Î¶¨Ï¶ò
    
    Parameters:
    - func: ÏµúÏ†ÅÌôîÌï† Ìï®Ïàò f(x, y)
    - x_bounds, y_bounds: ÌÉêÏÉâ ÏòÅÏó≠ Í≤ΩÍ≥Ñ
    - skip_pattern: Í±¥ÎÑàÎõ∏ Í∞ÑÍ≤©Ïùò Ìå®ÌÑ¥ (ÏÇ¨Ïù¥ÌÅ¥ Î∞òÎ≥µ)
    - initial_points: Í∞Å Ï∂ïÏùò Ï¥àÍ∏∞ ÌÉêÏÉâÏ†ê Ïàò
    - refinement_levels: Ï†ÅÏùëÏ†Å ÏÑ∏Î∞ÄÌôî Îã®Í≥Ñ Ïàò
    - refinement_factor: ÏÑ∏Î∞ÄÌôîÌï† ÏòÅÏó≠Ïùò ÌÅ¨Í∏∞ ÎπÑÏú®
    """
    
    start_time = time.time()
    
    # 1Îã®Í≥Ñ: Î∂àÍ∑†Îì± Í∞ÑÍ≤© Ï¥àÍ∏∞ ÌÉêÏÉâ
    if verbose:
        print("üîç 1Îã®Í≥Ñ: Î∂àÍ∑†Îì± Í∞ÑÍ≤© Ï¥àÍ∏∞ ÌÉêÏÉâ")
    
    x_min, x_max = x_bounds
    y_min, y_max = y_bounds
    
    # Î∂àÍ∑†Îì± Í∞ÑÍ≤©ÏúºÎ°ú Ï†êÎì§ ÏÉùÏÑ±
    x_points = []
    y_points = []
    
    # xÏ∂ï Î∂àÍ∑†Îì± Ï†êÎì§ ÏÉùÏÑ±
    pattern_idx = 0
    current_x = x_min
    while current_x <= x_max:
        x_points.append(current_x)
        step_size = (x_max - x_min) / initial_points * skip_pattern[pattern_idx % len(skip_pattern)]
        current_x += step_size
        pattern_idx += 1
    
    # yÏ∂ï Î∂àÍ∑†Îì± Ï†êÎì§ ÏÉùÏÑ± (Îã§Î•∏ Ìå®ÌÑ¥ ÏÇ¨Ïö©)
    pattern_idx = 1  # xÏôÄ Îã§Î•∏ ÏãúÏûëÏ†ê
    current_y = y_min
    while current_y <= y_max:
        y_points.append(current_y)
        step_size = (y_max - y_min) / initial_points * skip_pattern[pattern_idx % len(skip_pattern)]
        current_y += step_size
        pattern_idx += 1
    
    # Ï¥àÍ∏∞ Í∑∏Î¶¨Îìú ÌèâÍ∞Ä
    initial_evaluations = []
    for x in x_points:
        for y in y_points:
            value = func(x, y)
            initial_evaluations.append((x, y, value))
    
    # ÏµúÏÜüÍ∞í Ï∞æÍ∏∞
    best_point = min(initial_evaluations, key=lambda p: p[2])
    
    if verbose:
        print(f"   Ï¥àÍ∏∞ ÌÉêÏÉâÏ†ê Ïàò: {len(x_points)} √ó {len(y_points)} = {len(initial_evaluations)}")
        print(f"   Ï¥àÍ∏∞ ÏµúÏÜüÍ∞í: {best_point[2]:.6f} at ({best_point[0]:.4f}, {best_point[1]:.4f})")
    
    # 2Îã®Í≥Ñ: Ï†ÅÏùëÏ†Å ÏÑ∏Î∞ÄÌôî
    all_evaluations = initial_evaluations.copy()
    
    current_best = best_point
    current_x_bounds = x_bounds
    current_y_bounds = y_bounds
    
    for level in range(refinement_levels):
        if verbose:
            print(f"\nüéØ {level+2}Îã®Í≥Ñ: Ï†ÅÏùëÏ†Å ÏÑ∏Î∞ÄÌôî (Î†àÎ≤® {level+1})")
        
        # ÌòÑÏû¨ ÏµúÏÜüÍ∞í Ï£ºÎ≥Ä ÏòÅÏó≠ Ï†ïÏùò
        x_center, y_center = current_best[0], current_best[1]
        x_range = current_x_bounds[1] - current_x_bounds[0]
        y_range = current_y_bounds[1] - current_y_bounds[0]
        
        # ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠ Í≥ÑÏÇ∞
        refine_x_range = x_range * refinement_factor
        refine_y_range = y_range * refinement_factor
        
        new_x_min = max(x_bounds[0], x_center - refine_x_range/2)
        new_x_max = min(x_bounds[1], x_center + refine_x_range/2)
        new_y_min = max(y_bounds[0], y_center - refine_y_range/2)
        new_y_max = min(y_bounds[1], y_center + refine_y_range/2)
        
        # ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠ÏóêÏÑú Îçî Ï°∞Î∞ÄÌïú Í∑∏Î¶¨Îìú ÏÉùÏÑ±
        refine_points = initial_points // (level + 1) + 10  # Îã®Í≥ÑÎ≥ÑÎ°ú Ï†ê Ïàò Ï°∞Ï†ï
        
        # ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠Ïùò Î∂àÍ∑†Îì± Í∑∏Î¶¨Îìú
        refine_x_points = []
        refine_y_points = []
        
        # xÏ∂ï ÏÑ∏Î∞ÄÌôî Ï†êÎì§
        pattern_idx = level % len(skip_pattern)
        current_x = new_x_min
        while current_x <= new_x_max:
            refine_x_points.append(current_x)
            step_size = (new_x_max - new_x_min) / refine_points * skip_pattern[pattern_idx % len(skip_pattern)]
            current_x += step_size
            pattern_idx += 1
        
        # yÏ∂ï ÏÑ∏Î∞ÄÌôî Ï†êÎì§
        pattern_idx = (level + 2) % len(skip_pattern)
        current_y = new_y_min
        while current_y <= new_y_max:
            refine_y_points.append(current_y)
            step_size = (new_y_max - new_y_min) / refine_points * skip_pattern[pattern_idx % len(skip_pattern)]
            current_y += step_size
            pattern_idx += 1
        
        # ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠ ÌèâÍ∞Ä
        refine_evaluations = []
        for x in refine_x_points:
            for y in refine_y_points:
                # Ïù¥ÎØ∏ ÌèâÍ∞ÄÎêú Ï†êÎì§Í≥º ÎÑàÎ¨¥ Í∞ÄÍπåÏö¥ Ï†êÎì§ Ï†úÏô∏
                too_close = any(abs(x - ex) < 1e-6 and abs(y - ey) < 1e-6 
                               for ex, ey, _ in all_evaluations)
                if not too_close:
                    value = func(x, y)
                    refine_evaluations.append((x, y, value))
        
        all_evaluations.extend(refine_evaluations)
        
        # ÏÉàÎ°úÏö¥ ÏµúÏÜüÍ∞í Ï∞æÍ∏∞
        level_best = min(all_evaluations, key=lambda p: p[2])
        
        if verbose:
            print(f"   ÏÑ∏Î∞ÄÌôî ÏòÅÏó≠: [{new_x_min:.4f}, {new_x_max:.4f}] √ó [{new_y_min:.4f}, {new_y_max:.4f}]")
            print(f"   Ï∂îÍ∞Ä ÌÉêÏÉâÏ†ê: {len(refine_evaluations)}")
            print(f"   ÌòÑÏû¨ ÏµúÏÜüÍ∞í: {level_best[2]:.6f} at ({level_best[0]:.4f}, {level_best[1]:.4f})")
        
        current_best = level_best
        current_x_bounds = (new_x_min, new_x_max)
        current_y_bounds = (new_y_min, new_y_max)
    
    end_time = time.time()
    
    # Í≤∞Í≥º Ï†ïÎ¶¨
    final_best = min(all_evaluations, key=lambda p: p[2])
    
    results = {
        'minimum_value': final_best[2],
        'minimum_location': (final_best[0], final_best[1]),
        'total_evaluations': len(all_evaluations),
        'computation_time': end_time - start_time,
        'all_points': all_evaluations,
        'algorithm_params': {
            'skip_pattern': skip_pattern,
            'initial_points': initial_points,
            'refinement_levels': refinement_levels,
            'refinement_factor': refinement_factor
        }
    }
    
    if verbose:
        print(f"\n‚úÖ ÏµúÏ¢Ö Í≤∞Í≥º:")
        print(f"   ÏµúÏÜüÍ∞í: {final_best[2]:.8f}")
        print(f"   ÏúÑÏπò: ({final_best[0]:.6f}, {final_best[1]:.6f})")
        print(f"   Ï¥ù Ìï®Ïàò ÌèâÍ∞Ä ÌöüÏàò: {len(all_evaluations)}")
        print(f"   Í≥ÑÏÇ∞ ÏãúÍ∞Ñ: {end_time - start_time:.4f}Ï¥à")
    
    return results

# ÌÖåÏä§Ìä∏Ïö© Ìï®Ïàò Ï†ïÏùò
def test_function(x, y):
    """ÌÖåÏä§Ìä∏ Ìï®Ïàò: f(x,y) = cos(x) + cos(y)"""
    return np.cos(x) + np.cos(y)

print("üöÄ ÏÇ¨Ïö©Ïûê ÏßÄÏ†ï Í∑∏Î¶¨Îìú ÌÉêÏÉâ ÏïåÍ≥†Î¶¨Ï¶ò Íµ¨ÌòÑ ÏôÑÎ£å!")
print("Ïù¥Ï†ú ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÌÖåÏä§Ìä∏Ìï¥Î≥¥Í≤†ÏäµÎãàÎã§...")
````

````python
# Îã§ÏñëÌïú Ìå®ÌÑ¥ÏúºÎ°ú ÏïåÍ≥†Î¶¨Ï¶ò ÌÖåÏä§Ìä∏
print("\n" + "="*80)
print("üß™ Îã§ÏñëÌïú ÌÉêÏÉâ Ìå®ÌÑ¥ ÏÑ±Îä• ÎπÑÍµê")
print("="*80)

# ÌÖåÏä§Ìä∏Ìï† Îã§ÏñëÌïú Ìå®ÌÑ¥Îì§
test_patterns = [
    {
        'name': 'Í∑†Îì± Ìå®ÌÑ¥',
        'skip_pattern': [1, 1, 1, 1, 1],
        'refinement_levels': 3,
        'refinement_factor': 0.3
    },
    {
        'name': 'ÌîºÎ≥¥ÎÇòÏπò Ìå®ÌÑ¥', 
        'skip_pattern': [1, 1, 2, 3, 5],
        'refinement_levels': 3,
        'refinement_factor': 0.2
    },
    {
        'name': 'ÏßÄÏàòÏ†Å Ìå®ÌÑ¥',
        'skip_pattern': [1, 2, 4, 8, 4, 2],
        'refinement_levels': 4,
        'refinement_factor': 0.15
    },
    {
        'name': 'ÌååÎèô Ìå®ÌÑ¥',
        'skip_pattern': [1, 3, 5, 3, 1, 2],
        'refinement_levels': 4,
        'refinement_factor': 0.25
    }
]

comparison_results = []

for i, pattern_config in enumerate(test_patterns):
    print(f"\nüîπ {i+1}. {pattern_config['name']} ÌÖåÏä§Ìä∏")
    print(f"   Ìå®ÌÑ¥: {pattern_config['skip_pattern']}")
    
    result = custom_grid_search(
        func=test_function,
        x_bounds=(-2*np.pi, 2*np.pi),
        y_bounds=(-2*np.pi, 2*np.pi),
        skip_pattern=pattern_config['skip_pattern'],
        initial_points=35,
        refinement_levels=pattern_config['refinement_levels'],
        refinement_factor=pattern_config['refinement_factor'],
        verbose=False
    )
    
    comparison_results.append({
        'name': pattern_config['name'],
        'pattern': pattern_config['skip_pattern'],
        'minimum_value': result['minimum_value'],
        'location': result['minimum_location'],
        'evaluations': result['total_evaluations'],
        'time': result['computation_time'],
        'error': abs(result['minimum_value'] - (-2.0))
    })
    
    print(f"   ‚úÖ ÏµúÏÜüÍ∞í: {result['minimum_value']:.6f}")
    print(f"   üìç ÏúÑÏπò: ({result['minimum_location'][0]:.4f}, {result['minimum_location'][1]:.4f})")
    print(f"   üìä ÌèâÍ∞Ä ÌöüÏàò: {result['total_evaluations']}")
    print(f"   ‚è±Ô∏è ÏãúÍ∞Ñ: {result['computation_time']:.4f}Ï¥à")

# Í≤∞Í≥º ÎπÑÍµê ÌÖåÏù¥Î∏î
print(f"\n" + "="*80)
print("üìä Ìå®ÌÑ¥Î≥Ñ ÏÑ±Îä• ÎπÑÍµê ÏöîÏïΩ")
print("="*80)

header = f"{'Ìå®ÌÑ¥':<12} {'ÏµúÏÜüÍ∞í':<12} {'Ïò§Ï∞®':<12} {'ÌèâÍ∞ÄÌöüÏàò':<10} {'ÏãúÍ∞Ñ(Ï¥à)':<10} {'Ìö®Ïú®ÏÑ±':<10}"
print(header)
print("-" * len(header))

for result in comparison_results:
    efficiency = 1000 / result['evaluations']  # Îã®ÏúÑ ÌèâÍ∞ÄÎãπ Ìö®Ïú®ÏÑ± ÏßÄÌëú
    print(f"{result['name']:<12} {result['minimum_value']:<12.6f} {result['error']:<12.2e} "
          f"{result['evaluations']:<10} {result['time']:<10.4f} {efficiency:<10.2f}")

# ÏµúÍ≥† ÏÑ±Îä• Ìå®ÌÑ¥ ÏãùÎ≥Ñ
best_accuracy = min(comparison_results, key=lambda x: x['error'])
best_efficiency = min(comparison_results, key=lambda x: x['evaluations'])
best_speed = min(comparison_results, key=lambda x: x['time'])

print(f"\nüèÜ ÏµúÍ≥† ÏÑ±Îä•:")
print(f"‚Ä¢ ÏµúÍ≥† Ï†ïÌôïÎèÑ: {best_accuracy['name']} (Ïò§Ï∞®: {best_accuracy['error']:.2e})")
print(f"‚Ä¢ ÏµúÍ≥† Ìö®Ïú®ÏÑ±: {best_efficiency['name']} (ÌèâÍ∞Ä ÌöüÏàò: {best_efficiency['evaluations']})")
print(f"‚Ä¢ ÏµúÍ≥† ÏÜçÎèÑ: {best_speed['name']} (ÏãúÍ∞Ñ: {best_speed['time']:.4f}Ï¥à)")

# Ìå®ÌÑ¥Î≥Ñ ÏãúÍ∞ÅÌôî
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))
ensure_korean_font()

# ÌèâÍ∞Ä ÌöüÏàò ÎπÑÍµê
names = [r['name'] for r in comparison_results]
evaluations = [r['evaluations'] for r in comparison_results]
colors = ['skyblue', 'lightgreen', 'lightcoral', 'plum']

ax1.bar(names, evaluations, color=colors)
ax1.set_ylabel('Ìï®Ïàò ÌèâÍ∞Ä ÌöüÏàò')
ax1.set_title('Ìå®ÌÑ¥Î≥Ñ Í≥ÑÏÇ∞Îüâ ÎπÑÍµê')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True, alpha=0.3)

# Ïò§Ï∞® ÎπÑÍµê (Î°úÍ∑∏ Ïä§ÏºÄÏùº)
errors = [max(r['error'], 1e-10) for r in comparison_results]  # 0 Ïò§Ï∞® Î∞©ÏßÄ
ax2.bar(names, errors, color=colors)
ax2.set_ylabel('Ïò§Ï∞® (Î°úÍ∑∏ Ïä§ÏºÄÏùº)')
ax2.set_title('Ìå®ÌÑ¥Î≥Ñ Ï†ïÌôïÎèÑ ÎπÑÍµê')
ax2.set_yscale('log')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)

# Í≥ÑÏÇ∞ ÏãúÍ∞Ñ ÎπÑÍµê
times = [r['time'] for r in comparison_results]
ax3.bar(names, times, color=colors)
ax3.set_ylabel('Í≥ÑÏÇ∞ ÏãúÍ∞Ñ (Ï¥à)')
ax3.set_title('Ìå®ÌÑ¥Î≥Ñ ÏÜçÎèÑ ÎπÑÍµê')
ax3.tick_params(axis='x', rotation=45)
ax3.grid(True, alpha=0.3)

# Ìö®Ïú®ÏÑ± Ï¢ÖÌï© Ï†êÏàò (Ï†ïÍ∑úÌôîÎêú Î≥µÌï© ÏßÄÌëú)
max_eval = max(evaluations)
max_time = max(times)
max_error = max(errors) if max(errors) > 0 else 1

efficiency_scores = []
for r in comparison_results:
    # ÎÇÆÏùÑÏàòÎ°ù Ï¢ãÏùÄ ÏßÄÌëúÎì§ÏùÑ Ï†ïÍ∑úÌôî (0-1 Î≤îÏúÑ)
    eval_score = 1 - (r['evaluations'] / max_eval)
    time_score = 1 - (r['time'] / max_time)
    error_score = 1 - (r['error'] / max_error) if max_error > 0 else 1
    
    # Ï¢ÖÌï© Ï†êÏàò (Í∞ÄÏ§ë ÌèâÍ∑†)
    total_score = (eval_score * 0.4 + time_score * 0.3 + error_score * 0.3) * 100
    efficiency_scores.append(total_score)

ax4.bar(names, efficiency_scores, color=colors)
ax4.set_ylabel('Ï¢ÖÌï© Ìö®Ïú®ÏÑ± Ï†êÏàò')
ax4.set_title('Ìå®ÌÑ¥Î≥Ñ Ï¢ÖÌï© ÏÑ±Îä• Ï†êÏàò')
ax4.tick_params(axis='x', rotation=45)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nüí° Í≤∞Î°†:")
print(f"‚Ä¢ ÏÇ¨Ïö©Ïûê ÏßÄÏ†ï Í∑∏Î¶¨Îìú ÌÉêÏÉâ ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Îã§ÏñëÌïú Ìå®ÌÑ¥ÏóêÏÑú Î™®Îëê Ïö∞ÏàòÌïú ÏÑ±Îä•ÏùÑ Î≥¥ÏûÑ")
print(f"‚Ä¢ Î∂àÍ∑†Îì± Ìå®ÌÑ¥ÏùÄ Ìï®ÏàòÏùò ÌäπÏÑ±Ïóê ÎßûÍ≤å Ï°∞Ï†ï Í∞ÄÎä•")
print(f"‚Ä¢ Ï†ÅÏùëÏ†Å ÏÑ∏Î∞ÄÌôîÎ°ú Ï†ïÌôïÎèÑÏôÄ Ìö®Ïú®ÏÑ±Ïùò Í∑†Ìòï Îã¨ÏÑ±")
print(f"‚Ä¢ Î≥µÏû°Ìïú ÏµúÏ†ÅÌôî Î¨∏Ï†úÏóê Ïú†Ïó∞ÌïòÍ≤å Ï†ÅÏö© Í∞ÄÎä•Ìïú Î≤îÏö© ÏïåÍ≥†Î¶¨Ï¶ò")
````
