## 1. 최저점 구하기
$$ y = 2sin(x) + 0.5x , -5<=x<=13 $$

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
import math

# 한글 폰트 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

# Windows에서 한글 폰트 사용 (시스템에 따라 자동 선택)
import platform
if platform.system() == 'Windows':
    try:
        # Windows 한글 폰트 시도
        plt.rcParams['font.family'] = 'Malgun Gothic'
    except:
        try:
            plt.rcParams['font.family'] = 'AppleGothic'
        except:
            # 기본 한글 지원 폰트로 설정
            import matplotlib.font_manager as fm
            font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
            
            # 한글 폰트 찾기
            korean_fonts = []
            for font in font_list:
                try:
                    font_name = fm.FontProperties(fname=font).get_name()
                    if any(keyword in font_name.lower() for keyword in ['malgun', 'batang', 'dotum', 'gulim', 'arial']):
                        korean_fonts.append(font_name)
                except:
                    continue
            
            if korean_fonts:
                plt.rcParams['font.family'] = korean_fonts[0]
            else:
                # 마지막 수단: 시스템 기본 폰트
                plt.rcParams['font.family'] = ['sans-serif']
                
# 한글 폰트 확인
print(f"현재 설정된 폰트: {plt.rcParams['font.family']}")

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# x 범위를 0.001 간격으로 촘촘하게 설정
x = np.arange(-5, 13, 0.00001)

# 주어진 함수 y = 2sin(x) + 0.5x 계산
y = 2 * np.sin(x) + 0.5 * x

# 최솟값과 해당 x 값 찾기
min_index = np.argmin(y)
min_x = x[min_index]
min_y = y[min_index]

# 함수의 미분을 이용한 임계점 분석

# 함수 정의
def f(x):
    return 2 * np.sin(x) + 0.5 * x

# 1차 미분 함수 (도함수)
def f_prime(x):
    return 2 * np.cos(x) + 0.5

# 수치적 최솟값 찾기 (더 정확한 방법)
result = minimize_scalar(f, bounds=(-5, 13), method='bounded')
exact_min_x = result.x
exact_min_y = result.fun

# 임계점 찾기 (1차 미분 = 0인 점들)
# 2cos(x) + 0.5 = 0
# cos(x) = -0.25
critical_angles = []
x_test = np.linspace(-5, 13, 10000)
for i, x_val in enumerate(x_test[:-1]):
    if f_prime(x_val) * f_prime(x_test[i+1]) < 0:  # 부호가 바뀌는 지점
        critical_angles.append(x_val)

# 시각화를 위한 라이브러리 임포트
import matplotlib.pyplot as plt

# 한글 폰트 설정 (그래프에서 한글 깨짐 방지)
ensure_korean_font()

# 그래프 그리기
plt.figure(figsize=(12, 8))

# 함수 그래프 그리기
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')

# 최솟값 지점 표시
plt.plot(min_x, min_y, 'ro', markersize=10, label=f'최솟값: ({min_x:.3f}, {min_y:.3f})')

# 최솟값에서 수직선과 수평선 그리기
plt.axvline(x=min_x, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=min_y, color='red', linestyle='--', alpha=0.7)

# 그래프 꾸미기
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('함수 y = 2sin(x) + 0.5x의 그래프와 최솟값', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

# x축과 y축 범위 설정
plt.xlim(-5, 13)
plt.ylim(min_y - 1, max(y) + 1)

# 최솟값 정보를 텍스트로 표시
plt.text(min_x + 0.5, min_y + 0.5, 
         f'최솟값\nx = {min_x:.4f}\ny = {min_y:.4f}', 
         fontsize=10, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

plt.tight_layout()
plt.show()

# 수학적 분석 결과
print(f"\n=== 수학적 분석 결과 ===")
print(f"• 함수: y = 2sin(x) + 0.5x")
print(f"• 1차 미분: y' = 2cos(x) + 0.5")
print(f"• 임계점 조건: 2cos(x) + 0.5 = 0 → cos(x) = -0.25")
print(f"• cos(x) = -0.25의 해: x = ±arccos(0.25) + 2πn")
print(f"• 주어진 구간 [-5, 13]에서 임계점 개수: {len(critical_angles)}개")
print(f"• 전역 최솟값: x = {exact_min_x:.6f}, y = {exact_min_y:.6f}")

# arccos(-0.25) 계산
arccos_val = math.acos(0.25)
print(f"• arccos(0.25) = {arccos_val:.6f} (약 {math.degrees(arccos_val):.2f}도)")
print(f"• 임계점들은 π - arccos(0.25) + 2πn, π + arccos(0.25) + 2πn 형태")

# 상세한 시각화
plt.figure(figsize=(15, 10))

# 첫 번째 서브플롯: 원함수
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')
plt.plot(exact_min_x, exact_min_y, 'ro', markersize=10, label='최솟값')
for angle in critical_angles:
    plt.plot(angle, f(angle), 'go', markersize=8)
    
plt.xlabel('x')
plt.ylabel('y')
plt.title('원함수: y = 2sin(x) + 0.5x')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

# 두 번째 서브플롯: 1차 미분
plt.subplot(2, 2, 2)
y_prime = f_prime(x)
plt.plot(x, y_prime, 'g-', linewidth=2, label="y' = 2cos(x) + 0.5")
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='y = 0')
for angle in critical_angles:
    plt.plot(angle, 0, 'go', markersize=8)
    
plt.xlabel('x')
plt.ylabel("y'")
plt.title('1차 미분: 기울기의 변화')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

# 세 번째 서브플롯: 확대된 최솟값 근처
plt.subplot(2, 2, 3)
x_zoom = np.linspace(exact_min_x - 2, exact_min_x + 2, 1000)
y_zoom = f(x_zoom)
plt.plot(x_zoom, y_zoom, 'b-', linewidth=2)
plt.plot(exact_min_x, exact_min_y, 'ro', markersize=10)
plt.axvline(x=exact_min_x, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=exact_min_y, color='red', linestyle='--', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'최솟값 근처 확대\n(x={exact_min_x:.4f}, y={exact_min_y:.4f})')
plt.grid(True, alpha=0.3)

# 네 번째 서브플롯: 함수의 구성 요소 분해
plt.subplot(2, 2, 4)
sin_component = 2 * np.sin(x)
linear_component = 0.5 * x
plt.plot(x, sin_component, 'r--', linewidth=2, alpha=0.7, label='2sin(x)')
plt.plot(x, linear_component, 'g--', linewidth=2, alpha=0.7, label='0.5x')
plt.plot(x, y, 'b-', linewidth=2, label='y = 2sin(x) + 0.5x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('함수 구성 요소 분해')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-5, 13)

plt.tight_layout()
plt.show()

# 최솟값 정보 출력
print(f"=== 함수 y = 2sin(x) + 0.5x의 최솟값 분석 ===")
print(f"구간: [-5, 13]")
print(f"최솟값이 발생하는 x: {min_x:.6f}")
print(f"최솟값: {min_y:.6f}")
print(f"최솟값 위치: ({min_x:.4f}, {min_y:.4f})")

# 전체 구간에서 모든 극값들을 찾아 비교
from scipy.optimize import fminbound
import matplotlib.pyplot as plt

print("=== 전체 구간 [-5, 13]에서 모든 극값 분석 ===")

# 모든 임계점에서의 함수값 계산
all_critical_points = []

# 1차 미분이 0이 되는 점들을 더 정확히 찾기
x_fine = np.linspace(-5, 13, 50000)
y_prime_fine = 2 * np.cos(x_fine) + 0.5

# 부호가 바뀌는 지점들 찾기
for i in range(len(x_fine)-1):
    if y_prime_fine[i] * y_prime_fine[i+1] < 0:
        # 더 정확한 값을 위해 해당 구간에서 세분화
        x_local = np.linspace(x_fine[i], x_fine[i+1], 1000)
        y_prime_local = 2 * np.cos(x_local) + 0.5
        zero_idx = np.argmin(np.abs(y_prime_local))
        critical_x = x_local[zero_idx]
        critical_y = f(critical_x)
        all_critical_points.append((critical_x, critical_y))

# 구간 끝점들도 포함
all_critical_points.append((-5, f(-5)))
all_critical_points.append((13, f(13)))

# 정렬 및 출력
all_critical_points.sort()

print(f"발견된 모든 임계점 및 끝점들:")
for i, (x_val, y_val) in enumerate(all_critical_points):
    point_type = "끝점" if x_val == -5 or x_val == 13 else "임계점"
    print(f"{i+1}. x = {x_val:8.4f}, y = {y_val:8.4f} ({point_type})")

# 전역 최솟값과 최댓값 찾기
min_point = min(all_critical_points, key=lambda p: p[1])
max_point = max(all_critical_points, key=lambda p: p[1])

print(f"\n=== 최종 결과 ===")
print(f"🔵 전역 최솟값: x = {min_point[0]:.6f}, y = {min_point[1]:.6f}")
print(f"🔴 전역 최댓값: x = {max_point[0]:.6f}, y = {max_point[1]:.6f}")

# 시각화 개선
plt.figure(figsize=(14, 8))

# 함수 그래프
x_plot = np.linspace(-5, 13, 2000)
y_plot = 2 * np.sin(x_plot) + 0.5 * x_plot

plt.plot(x_plot, y_plot, 'b-', linewidth=2.5, label='y = 2sin(x) + 0.5x')

# 모든 임계점 표시
for x_val, y_val in all_critical_points:
    if x_val == -5 or x_val == 13:
        plt.plot(x_val, y_val, 'ks', markersize=8, label='끝점' if x_val == -5 else '')
    else:
        plt.plot(x_val, y_val, 'go', markersize=8, label='임계점' if x_val == all_critical_points[1][0] else '')

# 전역 최솟값과 최댓값 강조
plt.plot(min_point[0], min_point[1], 'ro', markersize=12, 
         label=f'전역 최솟값: ({min_point[0]:.3f}, {min_point[1]:.3f})')
plt.plot(max_point[0], max_point[1], 'mo', markersize=12, 
         label=f'전역 최댓값: ({max_point[0]:.3f}, {max_point[1]:.3f})')

# 최솟값에서 선 그리기
plt.axvline(x=min_point[0], color='red', linestyle='--', alpha=0.7)
plt.axhline(y=min_point[1], color='red', linestyle='--', alpha=0.7)

# 그래프 꾸미기
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('함수 y = 2sin(x) + 0.5x: 모든 극값 분석', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)
plt.xlim(-5, 13)

# 최솟값 정보 텍스트
plt.text(min_point[0] + 0.5, min_point[1] + 0.8, 
         f'전역 최솟값\nx = {min_point[0]:.4f}\ny = {min_point[1]:.4f}', 
         fontsize=11, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.4", facecolor="yellow", alpha=0.8))

plt.tight_layout()
plt.show()

# 수학적 분석
print(f"\n=== 수학적 해석 ===")
print(f"• 함수 y = 2sin(x) + 0.5x는 주기적 성분과 선형 성분의 합")
print(f"• 사인 함수의 진폭: 2")
print(f"• 선형 함수의 기울기: 0.5")
print(f"• 1차 미분: y' = 2cos(x) + 0.5")
print(f"• 임계점 조건: cos(x) = -0.25")
print(f"• cos(x) = -0.25는 약 {np.degrees(np.arccos(0.25)):.1f}°와 {360-np.degrees(np.arccos(0.25)):.1f}°에서 발생")
print(f"• 선형 성분 때문에 전체적으로 우상향하는 경향")
print(f"• 따라서 구간의 왼쪽 부분에서 최솟값이 발생")

# 한글 폰트 문제 해결을 위한 추가 설정
def setup_korean_font():
    """한글 폰트 설정 함수"""
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm
    import platform
    
    # 시스템별 한글 폰트 설정
    if platform.system() == 'Windows':
        # Windows 한글 폰트들
        fonts = ['Malgun Gothic', 'Batang', 'Dotum', 'Gulim', 'Arial Unicode MS']
    elif platform.system() == 'Darwin':  # macOS
        fonts = ['AppleGothic', 'Arial Unicode MS']
    else:  # Linux
        fonts = ['DejaVu Sans', 'Liberation Sans']
    
    # 사용 가능한 폰트 찾기
    available_fonts = [f.name for f in fm.fontManager.ttflist]
    
    for font in fonts:
        if font in available_fonts:
            plt.rcParams['font.family'] = font
            print(f"✅ 한글 폰트 설정 완료: {font}")
            break
    else:
        # 한글 폰트를 찾지 못한 경우
        print("⚠️ 한글 폰트를 찾지 못했습니다. 기본 설정을 사용합니다.")
        plt.rcParams['font.family'] = ['sans-serif']
    
    # 음수 기호 깨짐 방지
    plt.rcParams['axes.unicode_minus'] = False
    
    return plt.rcParams['font.family']

# 한글 폰트 설정 실행
current_font = setup_korean_font()

# 폰트 테스트
plt.figure(figsize=(8, 4))
plt.text(0.5, 0.5, '한글 폰트 테스트: 가나다라마바사아차카타파하', 
         ha='center', va='center', fontsize=16)
plt.text(0.5, 0.3, f'현재 폰트: {current_font}', 
         ha='center', va='center', fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title('한글 폰트 설정 확인')
plt.axis('off')
plt.tight_layout()
plt.show()
````

## 2. 숨겨진 최솟값 찾기

$$ f(x,y) = cos(x) + cos(y) $$     
$$-2{\pi} <= x <= 2{\pi},$$
$$-2{\pi} <= y <= 2{\pi} $$

````python
### 2번 문제 해결 접근 방법들

**문제**: $f(x,y) = \cos(x) + \cos(y)$에서 $-2\pi \leq x \leq 2\pi$, $-2\pi \leq y \leq 2\pi$ 범위의 최솟값 찾기

#### 🎯 **방법 1: 해석적 접근 (Analytical Method)**
- **편미분을 이용한 임계점 찾기**
  - $\frac{\partial f}{\partial x} = -\sin(x) = 0$ → $x = n\pi$ (정수 n)
  - $\frac{\partial f}{\partial y} = -\sin(y) = 0$ → $y = n\pi$ (정수 n)
  - 주어진 구간에서 가능한 임계점들: $(0,0), (\pi,0), (0,\pi), (\pi,\pi)$ 등
- **경계값 검사**
- **2차 미분 판정법으로 극값 분류**

#### 🔢 **방법 2: 수치적 최적화 (Numerical Optimization)**
- **scipy.optimize.minimize** 사용
- **Global optimization**: `scipy.optimize.differential_evolution`
- **Grid search + local optimization**

#### 📊 **방법 3: 격자점 탐색 (Grid Search)**
- **고해상도 격자로 전체 영역 스캔**
- **NumPy meshgrid 활용**
- **Brute force 방식으로 모든 점 검사**

#### 🎨 **방법 4: 시각화 기반 접근 (Visualization)**
- **3D Surface Plot**
- **Contour Plot (등고선 그래프)**
- **Heatmap**
- **시각적으로 최솟값 위치 확인 후 정밀 계산**

#### 🧮 **방법 5: 대칭성 활용 (Symmetry Analysis)**
- **함수의 주기성과 대칭성 이용**
- $\cos(x)$는 $2\pi$ 주기, 우함수
- $f(x,y) = f(-x,y) = f(x,-y) = f(-x,-y)$
- 첫 번째 사분면만 분석 후 확장

#### 🔬 **방법 6: 수학적 분석 (Mathematical Analysis)**
- **코사인 함수의 성질 활용**
- $\cos(x)$의 최솟값은 $x = \pi + 2n\pi$에서 -1
- $f(x,y) = \cos(x) + \cos(y)$의 최솟값은 $-1 + (-1) = -2$
- 최솟값 위치: $(x,y) = (\pi, \pi), (-\pi, \pi), (\pi, -\pi), (-\pi, -\pi)$

#### 🚀 **각 방법의 장단점**

| 방법 | 장점 | 단점 | 정확도 |
|------|------|------|--------|
| 해석적 | 정확한 해, 수학적 이해 | 복잡한 함수에 한계 | ⭐⭐⭐⭐⭐ |
| 수치적 최적화 | 범용성, 빠름 | 지역 최솟값 위험 | ⭐⭐⭐⭐ |
| 격자점 탐색 | 확실함, 직관적 | 계산량 많음 | ⭐⭐⭐⭐ |
| 시각화 | 직관적 이해 | 근사치 | ⭐⭐⭐ |
| 대칭성 활용 | 효율적 | 특수한 경우만 | ⭐⭐⭐⭐⭐ |
| 수학적 분석 | 가장 정확 | 함수별 분석 필요 | ⭐⭐⭐⭐⭐ |

#### 💡 **권장 접근 순서**
1. **수학적 분석**으로 이론적 답 구하기
2. **시각화**로 함수 형태 파악
3. **해석적 방법**으로 정확한 임계점 계산  
4. **수치적 방법**으로 검증
5. **모든 방법 결과 비교**
````

````python
# 2번 문제: f(x,y) = cos(x) + cos(y) 최솟값 찾기
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution
from mpl_toolkits.mplot3d import Axes3D

# 한글 폰트 설정
ensure_korean_font()

print("🎯 문제 2: f(x,y) = cos(x) + cos(y)의 최솟값 찾기")
print("📐 정의역: -2π ≤ x ≤ 2π, -2π ≤ y ≤ 2π")
print("="*60)

# 함수 정의
def f(x, y):
    """목적 함수: f(x,y) = cos(x) + cos(y)"""
    return np.cos(x) + np.cos(y)

def f_1d(vars):
    """1차원 배열 입력용 함수 (최적화 알고리즘용)"""
    return f(vars[0], vars[1])

# 정의역 설정
x_min, x_max = -2*np.pi, 2*np.pi
y_min, y_max = -2*np.pi, 2*np.pi

print("\n" + "="*60)
print("🧮 방법 1: 수학적 분석 (Mathematical Analysis)")
print("="*60)

print("\n📊 코사인 함수의 성질 분석:")
print("• cos(x)의 최솟값: -1 (x = π + 2nπ에서)")
print("• cos(x)의 최댓값: +1 (x = 2nπ에서)")
print("• f(x,y) = cos(x) + cos(y)이므로:")
print("  - 최솟값: (-1) + (-1) = -2")
print("  - 최댓값: (+1) + (+1) = +2")

print("\n🎯 최솟값이 발생하는 조건:")
print("• cos(x) = -1이고 cos(y) = -1일 때")
print("• x = π + 2nπ, y = π + 2mπ (n, m은 정수)")

print("\n📍 주어진 구간 [-2π, 2π]에서 최솟값 위치:")
theoretical_min_points = []
for x_val in [-np.pi, np.pi]:
    for y_val in [-np.pi, np.pi]:
        if x_min <= x_val <= x_max and y_min <= y_val <= y_max:
            theoretical_min_points.append((x_val, y_val))
            print(f"  • ({x_val/np.pi:.0f}π, {y_val/np.pi:.0f}π) = ({x_val:.4f}, {y_val:.4f})")
            print(f"    f({x_val:.4f}, {y_val:.4f}) = {f(x_val, y_val):.6f}")

theoretical_min_value = -2
print(f"\n✅ 이론적 최솟값: {theoretical_min_value}")

print("\n" + "="*60)
print("🔢 방법 2: 수치적 최적화 (Numerical Optimization)")
print("="*60)

# 방법 2-1: 지역 최적화 (여러 시작점에서)
print("\n🎯 방법 2-1: 여러 시작점에서 지역 최적화")
local_results = []
start_points = [
    (0, 0), (np.pi, 0), (0, np.pi), (np.pi, np.pi),
    (-np.pi, 0), (0, -np.pi), (-np.pi, -np.pi),
    (np.pi/2, np.pi/2), (-np.pi/2, -np.pi/2)
]

for i, (x0, y0) in enumerate(start_points):
    result = minimize(f_1d, [x0, y0], 
                     bounds=[(x_min, x_max), (y_min, y_max)],
                     method='L-BFGS-B')
    local_results.append(result)
    print(f"시작점 {i+1}: ({x0:.2f}, {y0:.2f}) → "
          f"최솟값 {result.fun:.6f}, 위치: ({result.x[0]:.4f}, {result.x[1]:.4f})")

# 최선의 지역 최적화 결과
best_local = min(local_results, key=lambda r: r.fun)
print(f"\n🏆 최선의 지역 최적화 결과: {best_local.fun:.6f}")
print(f"   위치: ({best_local.x[0]:.4f}, {best_local.x[1]:.4f})")

# 방법 2-2: 전역 최적화
print("\n🌐 방법 2-2: 전역 최적화 (Differential Evolution)")
bounds = [(x_min, x_max), (y_min, y_max)]
global_result = differential_evolution(f_1d, bounds, seed=42, maxiter=1000)
print(f"전역 최적화 결과: {global_result.fun:.6f}")
print(f"위치: ({global_result.x[0]:.4f}, {global_result.x[1]:.4f})")

print("\n" + "="*60)
print("📊 방법 3: 고해상도 격자 탐색 (High-Resolution Grid Search)")
print("="*60)

# 고해상도 격자 생성
resolution = 1000  # 격자 해상도
x_grid = np.linspace(x_min, x_max, resolution)
y_grid = np.linspace(y_min, y_max, resolution)
X, Y = np.meshgrid(x_grid, y_grid)
Z = f(X, Y)

# 최솟값과 위치 찾기
min_idx = np.unravel_index(np.argmin(Z), Z.shape)
grid_min_x = X[min_idx]
grid_min_y = Y[min_idx]
grid_min_value = Z[min_idx]

print(f"격자 해상도: {resolution} × {resolution} = {resolution**2:,} 점")
print(f"격자 탐색 최솟값: {grid_min_value:.6f}")
print(f"위치: ({grid_min_x:.4f}, {grid_min_y:.4f})")

# 최솟값 근처의 다른 점들도 확인
min_threshold = grid_min_value + 1e-10
min_mask = Z <= min_threshold
min_indices = np.where(min_mask)
print(f"\n최솟값 근처 점들의 개수: {len(min_indices[0])}")

print("\n" + "="*60)
print("📈 방법 4: 해석적 접근 (Critical Points Analysis)")
print("="*60)

print("\n🔍 편미분을 이용한 임계점 찾기:")
print("∂f/∂x = -sin(x) = 0  →  x = nπ")
print("∂f/∂y = -sin(y) = 0  →  y = mπ")

# 주어진 구간에서 가능한 임계점들
critical_points = []
for n in range(-2, 3):  # -2π부터 2π까지
    for m in range(-2, 3):
        x_crit = n * np.pi
        y_crit = m * np.pi
        if x_min <= x_crit <= x_max and y_min <= y_crit <= y_max:
            f_val = f(x_crit, y_crit)
            critical_points.append((x_crit, y_crit, f_val))

# 임계점들 정렬 (함수값 기준)
critical_points.sort(key=lambda p: p[2])

print(f"\n📍 발견된 임계점들 (총 {len(critical_points)}개):")
for i, (x, y, fval) in enumerate(critical_points):
    point_type = "최솟값" if abs(fval - (-2)) < 1e-10 else "최댓값" if abs(fval - 2) < 1e-10 else "안장점"
    print(f"{i+1:2d}. ({x/np.pi:+.0f}π, {y/np.pi:+.0f}π) = ({x:+.4f}, {y:+.4f}), "
          f"f = {fval:+.6f} ({point_type})")

# 최솟값들만 따로 추출
min_points = [p for p in critical_points if abs(p[2] - (-2)) < 1e-10]
print(f"\n🎯 최솟값 위치들 (총 {len(min_points)}개):")
for i, (x, y, fval) in enumerate(min_points):
    print(f"{i+1}. ({x/np.pi:+.0f}π, {y/np.pi:+.0f}π) = ({x:+.4f}, {y:+.4f})")

print("\n" + "="*60)
print("📊 결과 종합 및 검증")
print("="*60)

results_summary = {
    "이론적 분석": theoretical_min_value,
    "최선 지역 최적화": best_local.fun,
    "전역 최적화": global_result.fun,
    "격자 탐색": grid_min_value,
    "해석적 임계점": min_points[0][2] if min_points else None
}

print("\n🏆 각 방법별 최솟값 결과:")
for method, value in results_summary.items():
    if value is not None:
        error = abs(value - theoretical_min_value)
        print(f"• {method:15s}: {value:10.6f} (오차: {error:.2e})")

print(f"\n✅ 최종 답: f(x,y) = cos(x) + cos(y)의 최솟값은 {theoretical_min_value}")
print(f"📍 최솟값 위치: (±π, ±π)의 조합 (총 4개 점)")
print(f"🎯 정확한 위치들:")
for x, y, _ in min_points:
    print(f"   • ({x:+.4f}, {y:+.4f})")

# 오차 분석
print(f"\n🔬 수치적 방법들의 정확도:")
numerical_methods = ["최선 지역 최적화", "전역 최적화", "격자 탐색"]
for method in numerical_methods:
    if method in results_summary:
        error = abs(results_summary[method] - theoretical_min_value)
        accuracy = max(0, -np.log10(error)) if error > 0 else float('inf')
        print(f"• {method}: 오차 {error:.2e}, 정확도 ~{accuracy:.1f} 자릿수")
````

## 🧠 사용자 지정 그리드 탐색 알고리즘

### 알고리즘 특징:
1. **불균등 간격 탐색**: 정해진 패턴으로 간격을 건너뛰며 탐색
2. **적응적 세밀화**: 최솟값 발견 시 주변 영역을 더 세밀하게 탐색
3. **계층적 접근**: 거친 탐색 → 세밀한 탐색의 다단계 과정
4. **효율적 커버리지**: 전체 영역을 효과적으로 커버하면서도 계산량 최적화

이 알고리즘은 일반적인 균등 그리드 탐색보다 효율적이면서도, 복잡한 함수의 숨겨진 최솟값을 찾아내는 능력이 뛰어납니다.

````python
import numpy as np
import matplotlib.pyplot as plt
from typing import Callable, Tuple, List, Dict
import time

def custom_grid_search(func: Callable, 
                      x_bounds: Tuple[float, float], 
                      y_bounds: Tuple[float, float],
                      skip_pattern: List[int] = [1, 2, 3, 2, 1],  # 건너뛸 간격 패턴
                      initial_points: int = 50,  # 초기 탐색점 수
                      refinement_levels: int = 3,  # 세밀화 단계 수
                      refinement_factor: float = 0.2,  # 세밀화 영역 비율
                      verbose: bool = True) -> Dict:
    """
    사용자 지정 그리드 탐색 알고리즘
    
    Parameters:
    - func: 최적화할 함수 f(x, y)
    - x_bounds, y_bounds: 탐색 영역 경계
    - skip_pattern: 건너뛸 간격의 패턴 (사이클 반복)
    - initial_points: 각 축의 초기 탐색점 수
    - refinement_levels: 적응적 세밀화 단계 수
    - refinement_factor: 세밀화할 영역의 크기 비율
    """
    
    start_time = time.time()
    
    # 1단계: 불균등 간격 초기 탐색
    if verbose:
        print("🔍 1단계: 불균등 간격 초기 탐색")
    
    x_min, x_max = x_bounds
    y_min, y_max = y_bounds
    
    # 불균등 간격으로 점들 생성
    x_points = []
    y_points = []
    
    # x축 불균등 점들 생성
    pattern_idx = 0
    current_x = x_min
    while current_x <= x_max:
        x_points.append(current_x)
        step_size = (x_max - x_min) / initial_points * skip_pattern[pattern_idx % len(skip_pattern)]
        current_x += step_size
        pattern_idx += 1
    
    # y축 불균등 점들 생성 (다른 패턴 사용)
    pattern_idx = 1  # x와 다른 시작점
    current_y = y_min
    while current_y <= y_max:
        y_points.append(current_y)
        step_size = (y_max - y_min) / initial_points * skip_pattern[pattern_idx % len(skip_pattern)]
        current_y += step_size
        pattern_idx += 1
    
    # 초기 그리드 평가
    initial_evaluations = []
    for x in x_points:
        for y in y_points:
            value = func(x, y)
            initial_evaluations.append((x, y, value))
    
    # 최솟값 찾기
    best_point = min(initial_evaluations, key=lambda p: p[2])
    
    if verbose:
        print(f"   초기 탐색점 수: {len(x_points)} × {len(y_points)} = {len(initial_evaluations)}")
        print(f"   초기 최솟값: {best_point[2]:.6f} at ({best_point[0]:.4f}, {best_point[1]:.4f})")
    
    # 2단계: 적응적 세밀화
    all_evaluations = initial_evaluations.copy()
    
    current_best = best_point
    current_x_bounds = x_bounds
    current_y_bounds = y_bounds
    
    for level in range(refinement_levels):
        if verbose:
            print(f"\n🎯 {level+2}단계: 적응적 세밀화 (레벨 {level+1})")
        
        # 현재 최솟값 주변 영역 정의
        x_center, y_center = current_best[0], current_best[1]
        x_range = current_x_bounds[1] - current_x_bounds[0]
        y_range = current_y_bounds[1] - current_y_bounds[0]
        
        # 세밀화 영역 계산
        refine_x_range = x_range * refinement_factor
        refine_y_range = y_range * refinement_factor
        
        new_x_min = max(x_bounds[0], x_center - refine_x_range/2)
        new_x_max = min(x_bounds[1], x_center + refine_x_range/2)
        new_y_min = max(y_bounds[0], y_center - refine_y_range/2)
        new_y_max = min(y_bounds[1], y_center + refine_y_range/2)
        
        # 세밀화 영역에서 더 조밀한 그리드 생성
        refine_points = initial_points // (level + 1) + 10  # 단계별로 점 수 조정
        
        # 세밀화 영역의 불균등 그리드
        refine_x_points = []
        refine_y_points = []
        
        # x축 세밀화 점들
        pattern_idx = level % len(skip_pattern)
        current_x = new_x_min
        while current_x <= new_x_max:
            refine_x_points.append(current_x)
            step_size = (new_x_max - new_x_min) / refine_points * skip_pattern[pattern_idx % len(skip_pattern)]
            current_x += step_size
            pattern_idx += 1
        
        # y축 세밀화 점들
        pattern_idx = (level + 2) % len(skip_pattern)
        current_y = new_y_min
        while current_y <= new_y_max:
            refine_y_points.append(current_y)
            step_size = (new_y_max - new_y_min) / refine_points * skip_pattern[pattern_idx % len(skip_pattern)]
            current_y += step_size
            pattern_idx += 1
        
        # 세밀화 영역 평가
        refine_evaluations = []
        for x in refine_x_points:
            for y in refine_y_points:
                # 이미 평가된 점들과 너무 가까운 점들 제외
                too_close = any(abs(x - ex) < 1e-6 and abs(y - ey) < 1e-6 
                               for ex, ey, _ in all_evaluations)
                if not too_close:
                    value = func(x, y)
                    refine_evaluations.append((x, y, value))
        
        all_evaluations.extend(refine_evaluations)
        
        # 새로운 최솟값 찾기
        level_best = min(all_evaluations, key=lambda p: p[2])
        
        if verbose:
            print(f"   세밀화 영역: [{new_x_min:.4f}, {new_x_max:.4f}] × [{new_y_min:.4f}, {new_y_max:.4f}]")
            print(f"   추가 탐색점: {len(refine_evaluations)}")
            print(f"   현재 최솟값: {level_best[2]:.6f} at ({level_best[0]:.4f}, {level_best[1]:.4f})")
        
        current_best = level_best
        current_x_bounds = (new_x_min, new_x_max)
        current_y_bounds = (new_y_min, new_y_max)
    
    end_time = time.time()
    
    # 결과 정리
    final_best = min(all_evaluations, key=lambda p: p[2])
    
    results = {
        'minimum_value': final_best[2],
        'minimum_location': (final_best[0], final_best[1]),
        'total_evaluations': len(all_evaluations),
        'computation_time': end_time - start_time,
        'all_points': all_evaluations,
        'algorithm_params': {
            'skip_pattern': skip_pattern,
            'initial_points': initial_points,
            'refinement_levels': refinement_levels,
            'refinement_factor': refinement_factor
        }
    }
    
    if verbose:
        print(f"\n✅ 최종 결과:")
        print(f"   최솟값: {final_best[2]:.8f}")
        print(f"   위치: ({final_best[0]:.6f}, {final_best[1]:.6f})")
        print(f"   총 함수 평가 횟수: {len(all_evaluations)}")
        print(f"   계산 시간: {end_time - start_time:.4f}초")
    
    return results

# 테스트용 함수 정의
def test_function(x, y):
    """테스트 함수: f(x,y) = cos(x) + cos(y)"""
    return np.cos(x) + np.cos(y)

print("🚀 사용자 지정 그리드 탐색 알고리즘 구현 완료!")
print("이제 알고리즘을 테스트해보겠습니다...")
````
